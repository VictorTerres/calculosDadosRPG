```python
#Valores em comum para todos os Testes

import random
import math

valor_ac = 15
modificador_atributo = 5
modificador_acerto_proficiencia = 2
modificador_acerto_extra = 2
vantagem_grito = 0
vantagem_vex = 0
rerolar_espada = 0
```

```python
## Ataque do Chakram Duplo - Inimigo Único (Monstruosidade) - Estilo: Arquearia

# Criar vetores para armazenar as rolagens e os ataques
rolagens = []
ataques = []

################################# Turno 01 - Marca do Caçador + Ataque Principal
# Ataque Principal
rolagem_1a = random.randint(1, 20)

if rolagem_1a == 20: #Caso o ataque seja um crítico!
  rolagem_1a += modificador_atributo
  rolagem_1a += modificador_acerto_proficiencia
  rolagem_1a += modificador_acerto_extra

  rolagens.append(rolagem_1a)

  dano_chakram_1a = random.randint(1, 6) + random.randint(1, 6)
  dano_marca_1a = random.randint(1, 6) + random.randint(1, 6)
  dano_final_1a = dano_chakram_1a + dano_marca_1a + modificador_atributo
  ataques.append(dano_final_1a)
elif rolagem_1a + modificador_atributo + modificador_acerto_proficiencia + modificador_acerto_extra >= valor_ac: # Caso acerte (não crítico)
  rolagem_1a += modificador_atributo
  rolagem_1a += modificador_acerto_proficiencia
  rolagem_1a += modificador_acerto_extra
  rolagens.append(rolagem_1a)

  dano_chakram_1a = random.randint(1, 6)
  dano_marca_1a = random.randint(1, 6)
  dano_final_1a = dano_chakram_1a + dano_marca_1a + modificador_atributo
  ataques.append(dano_final_1a)
else: # Caso não acerte
  rolagem_1a += modificador_atributo
  rolagem_1a += modificador_acerto_proficiencia
  rolagem_1a += modificador_acerto_extra
  rolagens.append(rolagem_1a)
  dano_chakram_1a = 0
  dano_marca_1a = 0
  dano_final_1a = 0
  ataques.append(dano_final_1a)

#print(f'Turno 1: Ataque Principal {rolagem_1a} com dano: (Chakram 1) {dano_chakram_1a} + (Marca) {dano_marca_1a} + (Modificador) {modificador_atributo} = {dano_final_1a}')

################################## Turno 02 - Grito de Guerra + Ataque Principal
# Ataque Principal
vantagem_grito = 1

if vantagem_grito or vantagem_vex:
  rolagem_2a = max(random.randint(1, 20), random.randint(1, 20))
else:
  rolagem_2a = random.randint(1, 20)

if rolagem_2a == 20: #Caso o ataque seja um crítico!
  rolagem_2a += modificador_atributo
  rolagem_2a += modificador_acerto_proficiencia
  rolagem_2a += modificador_acerto_extra

  rolagens.append(rolagem_2a)

  dano_chakram_2a = random.randint(1, 6) + random.randint(1, 6)
  dano_marca_2a = random.randint(1, 6) + random.randint(1, 6)
  dano_final_2a = dano_chakram_2a + dano_marca_2a + modificador_atributo
  ataques.append(dano_final_2a)
  vantagem_vex = 1
elif rolagem_2a + modificador_atributo + modificador_acerto_proficiencia + modificador_acerto_extra >= valor_ac: # Caso acerte (não crítico)
  rolagem_2a += modificador_atributo
  rolagem_2a += modificador_acerto_proficiencia
  rolagem_2a += modificador_acerto_extra
  rolagens.append(rolagem_2a)

  dano_chakram_2a = random.randint(1, 6)
  dano_marca_2a = random.randint(1, 6)
  dano_final_2a = dano_chakram_2a + dano_marca_2a + modificador_atributo
  ataques.append(dano_final_2a)
  vantagem_vex = 1
else: # Caso não acerte
  rolagem_2a += modificador_atributo
  rolagem_2a += modificador_acerto_proficiencia
  rolagem_2a += modificador_acerto_extra
  rolagens.append(rolagem_2a)
  dano_chakram_2a = 0
  dano_marca_2a = 0
  dano_final_2a = 0
  ataques.append(dano_final_2a)
  vantagem_vex = 0
  levou_dano = random.randint(1, 1000)

  if levou_dano <= 77:
    vantagem_grito = 0
#print(f'Turno 2: Ataque Principal {rolagem_2a} com dano: (Chakram) {dano_chakram_2a} + (Marca) {dano_marca_2a} + (Modificador) {modificador_atributo} = {dano_final_2a}')

################################# Turno 03 ao 10 - Ataque Principal e Secundário
for i in range(3, 11): # Loop para os Ataques
    # Ataque Principal
    if vantagem_grito or vantagem_vex:
      rolagem_pa = max(random.randint(1, 20), random.randint(1, 20))
    else:
      rolagem_pa = random.randint(1, 20)

    if rolagem_pa == 20: #Caso o ataque principal seja um crítico!
      rolagem_pa += modificador_atributo
      rolagem_pa += modificador_acerto_proficiencia
      rolagem_pa += modificador_acerto_extra

      rolagens.append(rolagem_pa)

      dano_chakram1 = random.randint(1, 6) + random.randint(1, 6)
      dano_marca1 = random.randint(1, 6) + random.randint(1, 6)
      dano_final_pa = dano_chakram1 + dano_marca1 + modificador_atributo
      vantagem_vex = 1
    elif rolagem_pa + modificador_atributo + modificador_acerto_proficiencia + modificador_acerto_extra >= valor_ac: # Caso acerte (não crítico)
      rolagem_pa += modificador_atributo
      rolagem_pa += modificador_acerto_proficiencia
      rolagem_pa += modificador_acerto_extra
      rolagens.append(rolagem_pa)

      dano_chakram1 = random.randint(1, 6)
      dano_marca1 = random.randint(1, 6)
      dano_final_pa = dano_chakram1 + dano_marca1 + modificador_atributo
      vantagem_vex = 1
    else: # Caso não acerte
      rolagem_pa += modificador_atributo
      rolagem_pa += modificador_acerto_proficiencia
      rolagens.append(rolagem_pa)
      dano_chakram1 = 0
      dano_marca1 = 0
      dano_final_pa = 0
      vantagem_vex = 0

    # Ataque Secundário
    if vantagem_grito or vantagem_vex:
      rolagem_sa = max(random.randint(1, 20), random.randint(1, 20))
    else:
      rolagem_sa = random.randint(1, 20)

    if rolagem_sa == 20: #Caso o ataque secundário seja um crítico!
        rolagem_sa += modificador_atributo
        rolagem_sa += modificador_acerto_proficiencia
        rolagem_sa += modificador_acerto_extra

        rolagens.append(rolagem_sa)

        dano_chakram2 = random.randint(1, 6) + random.randint(1, 6)
        dano_marca2 = random.randint(1, 6) + random.randint(1, 6)
        dano_final_sa = dano_chakram2 + dano_marca2
        vantagem_vex = 1
    elif rolagem_sa + modificador_atributo + modificador_acerto_proficiencia + modificador_acerto_extra >= valor_ac: # Caso acerte (não crítico)
        rolagem_sa += modificador_atributo
        rolagem_sa += modificador_acerto_proficiencia
        rolagem_sa += modificador_acerto_extra

        rolagens.append(rolagem_sa)

        dano_chakram2 = random.randint(1, 6)
        dano_marca2 = random.randint(1, 6)
        dano_final_sa = dano_chakram2 + dano_marca2
        vantagem_vex = 1
    else: # Caso não acerte
        rolagem_sa += modificador_atributo
        rolagem_sa += modificador_acerto_proficiencia
        rolagem_sa += modificador_acerto_extra
        rolagens.append(rolagem_sa)
        dano_chakram2 = 0
        dano_marca2 = 0
        dano_final_sa = 0
        vantagem_vex = 0

    #Soma dos dois ataques para cálculo de média de dano por Rodada
    dano_final = dano_final_pa + dano_final_sa
    ataques.append(dano_final)

    #print(f'Turno {i}: Ataque Principal {rolagem_pa} com dano: (Chakram 1) {dano_chakram1} + (Marca) {dano_marca1} + (Modificador) {modificador_atributo} = {dano_final_pa}')
    #print(f'Turno {i}: Ataque Secundário {rolagem_sa} com dano: (Chakram 2) {dano_chakram2} + (Marca) {dano_marca2} = {dano_final_sa} | Dano Total: {dano_final}')

    #Reset dos danos para não cumular
    dano_chakram1 = 0
    dano_chakram2 = 0
    dano_marca1 = 0
    dano_marca2 = 0
    dano_final_pa = 0
    dano_final_sa = 0
    dano_final = 0
    levou_dano = random.randint(1, 1000)

    if levou_dano <= 77:
      vantagem_grito = 0

################################################################################
# Calcular a média dos resultados
media_rolagens = sum(rolagens) / len(rolagens)
media_dano = sum(ataques) / len(ataques)

# Formatar os vetores para exibição com 2 dígitos
rolagens_formatadas = [f"{rolagem:02d}" for rolagem in rolagens]
ataques_formatados = [f"{ataque:02d}" for ataque in ataques]

# Exibir a média dos resultados e os vetores formatados
print(f"A média das rolagens é: {media_rolagens}")
print(rolagens_formatadas)
print(f"A média dos danos: {media_dano}")
print(ataques_formatados)
```

```python
import random
import math

# Initialize lists to store the averages from each run
all_avg_rolagens = []
all_avg_dano = []
all_advantage_attacks = [] # List to store whether an attack had advantage

# Run the simulation 100000 times
for _ in range(100000):
    # Criar vetores para armazenar as rolagens e os ataques
    rolagens = []
    ataques = []
    advantage_attacks_this_run = [] # List to store advantage status for this run

    ################################# Turno 01 - Marca do Caçador + Ataque Principal
    # Ataque Principal
    rolagem_1a = random.randint(1, 20)
    advantage_attacks_this_run.append(0) # No advantage in Turn 1

    if rolagem_1a == 20: #Caso o ataque seja um crítico!
      rolagem_1a += modificador_atributo
      rolagem_1a += modificador_acerto_proficiencia
      rolagem_1a += modificador_acerto_extra

      rolagens.append(rolagem_1a)

      dano_chakram_1a = random.randint(1, 6) + random.randint(1, 6)
      dano_marca_1a = random.randint(1, 6) + random.randint(1, 6)
      dano_final_1a = dano_chakram_1a + dano_marca_1a + modificador_atributo
      ataques.append(dano_final_1a)
    elif rolagem_1a + modificador_atributo + modificador_acerto_proficiencia + modificador_acerto_extra >= valor_ac: # Caso acerte (não crítico)
      rolagem_1a += modificador_atributo
      rolagem_1a += modificador_acerto_proficiencia
      rolagem_1a += modificador_acerto_extra
      rolagens.append(rolagem_1a)

      dano_chakram_1a = random.randint(1, 6)
      dano_marca_1a = random.randint(1, 6)
      dano_final_1a = dano_chakram_1a + dano_marca_1a + modificador_atributo
      ataques.append(dano_final_1a)
    else: # Caso não acerte
      rolagem_1a += modificador_atributo
      rolagem_1a += modificador_acerto_proficiencia
      rolagem_1a += modificador_acerto_extra
      rolagens.append(rolagem_1a)
      dano_chakram_1a = 0
      dano_marca_1a = 0
      dano_final_1a = 0
      ataques.append(dano_final_1a)

    #print(f'Turno 1: Ataque Principal {rolagem_1a} com dano: (Chakram 1) {dano_chakram_1a} + (Marca) {dano_marca_1a} + (Modificador) {modificador_atributo} = {dano_final_1a}')

    ################################## Turno 02 - Grito de Guerra + Ataque Principal
    # Ataque Principal
    vantagem_grito = 1

    if vantagem_grito or vantagem_vex:
      rolagem_2a = max(random.randint(1, 20), random.randint(1, 20))
      advantage_attacks_this_run.append(1) # Advantage in Turn 2
    else:
      rolagem_2a = random.randint(1, 20)
      advantage_attacks_this_run.append(0) # No advantage in Turn 2


    if rolagem_2a == 20: #Caso o ataque seja um crítico!
      rolagem_2a += modificador_atributo
      rolagem_2a += modificador_acerto_proficiencia
      rolagem_2a += modificador_acerto_extra

      rolagens.append(rolagem_2a)

      dano_chakram_2a = random.randint(1, 6) + random.randint(1, 6)
      dano_marca_2a = random.randint(1, 6) + random.randint(1, 6)
      dano_final_2a = dano_chakram_2a + dano_marca_2a + modificador_atributo
      ataques.append(dano_final_2a)
      vantagem_vex = 1
    elif rolagem_2a + modificador_atributo + modificador_acerto_proficiencia + modificador_acerto_extra >= valor_ac: # Caso acerte (não crítico)
      rolagem_2a += modificador_atributo
      rolagem_2a += modificador_acerto_proficiencia
      rolagem_2a += modificador_acerto_extra
      rolagens.append(rolagem_2a)

      dano_chakram_2a = random.randint(1, 6)
      dano_marca_2a = random.randint(1, 6)
      dano_final_2a = dano_chakram_2a + dano_marca_2a + modificador_atributo
      ataques.append(dano_final_2a)
      vantagem_vex = 1
    else: # Caso não acerte
      rolagem_2a += modificador_atributo
      rolagem_2a += modificador_acerto_proficiencia
      rolagem_2a += modificador_acerto_extra
      rolagens.append(rolagem_2a)
      dano_chakram_2a = 0
      dano_marca_2a = 0
      dano_final_2a = 0
      ataques.append(dano_final_2a)
      vantagem_vex = 0
      levou_dano = random.randint(1, 1000)

      if levou_dano <= 77:
        vantagem_grito = 0
    #print(f'Turno 2: Ataque Principal {rolagem_2a} com dano: (Chakram) {dano_chakram_2a} + (Marca) {dano_marca_2a} + (Modificador) {modificador_atributo} = {dano_final_2a}')

    ################################# Turno 03 ao 10 - Ataque Principal e Secundário
    for i in range(3, 11): # Loop para os Ataques
        # Ataque Principal
        if vantagem_grito or vantagem_vex:
          rolagem_pa = max(random.randint(1, 20), random.randint(1, 20))
          advantage_attacks_this_run.append(1) # Advantage for principal attack
        else:
          rolagem_pa = random.randint(1, 20)
          advantage_attacks_this_run.append(0) # No advantage for principal attack


        if rolagem_pa == 20: #Caso o ataque principal seja um crítico!
          rolagem_pa += modificador_atributo
          rolagem_pa += modificador_acerto_proficiencia
          rolagem_pa += modificador_acerto_extra

          rolagens.append(rolagem_pa)

          dano_chakram1 = random.randint(1, 6) + random.randint(1, 6)
          dano_marca1 = random.randint(1, 6) + random.randint(1, 6)
          dano_final_pa = dano_chakram1 + dano_marca1 + modificador_atributo
          vantagem_vex = 1
        elif rolagem_pa + modificador_atributo + modificador_acerto_proficiencia + modificador_acerto_extra >= valor_ac: # Caso acerte (não crítico)
          rolagem_pa += modificador_atributo
          rolagem_pa += modificador_acerto_proficiencia
          rolagem_pa += modificador_acerto_extra
          rolagens.append(rolagem_pa)

          dano_chakram1 = random.randint(1, 6)
          dano_marca1 = random.randint(1, 6)
          dano_final_pa = dano_chakram1 + dano_marca1 + modificador_atributo
          vantagem_vex = 1
        else: # Caso não acerte
          rolagem_pa += modificador_atributo
          rolagem_pa += modificador_acerto_proficiencia
          rolagens.append(rolagem_pa)
          dano_chakram1 = 0
          dano_marca1 = 0
          dano_final_pa = 0
          vantagem_vex = 0

        # Ataque Secundário
        if vantagem_grito or vantagem_vex:
          rolagem_sa = max(random.randint(1, 20), random.randint(1, 20))
          advantage_attacks_this_run.append(1) # Advantage for secondary attack
        else:
          rolagem_sa = random.randint(1, 20)
          advantage_attacks_this_run.append(0) # No advantage for secondary attack


        if rolagem_sa == 20: #Caso o ataque secundário seja um crítico!
            rolagem_sa += modificador_atributo
            rolagem_sa += modificador_acerto_proficiencia
            rolagem_sa += modificador_acerto_extra

            rolagens.append(rolagem_sa)

            dano_chakram2 = random.randint(1, 6) + random.randint(1, 6)
            dano_marca2 = random.randint(1, 6) + random.randint(1, 6)
            dano_final_sa = dano_chakram2 + dano_marca2
            vantagem_vex = 1
        elif rolagem_sa + modificador_atributo + modificador_acerto_proficiencia + modificador_acerto_extra >= valor_ac: # Caso acerte (não crítico)
            rolagem_sa += modificador_atributo
            rolagem_sa += modificador_acerto_proficiencia
            rolagem_sa += modificador_acerto_extra

            rolagens.append(rolagem_sa)

            dano_chakram2 = random.randint(1, 6)
            dano_marca2 = random.randint(1, 6)
            dano_final_sa = dano_chakram2 + dano_marca2
            vantagem_vex = 1
        else: # Caso não acerte
            rolagem_sa += modificador_atributo
            rolagem_sa += modificador_acerto_proficiencia
            rolagem_sa += modificador_acerto_extra
            rolagens.append(rolagem_sa)
            dano_chakram2 = 0
            dano_marca2 = 0
            dano_final_sa = 0
            vantagem_vex = 0

        #Soma dos dois ataques para cálculo de média de dano por Rodada
        dano_final = dano_final_pa + dano_final_sa
        ataques.append(dano_final)

        #print(f'Turno {i}: Ataque Principal {rolagem_pa} com dano: (Chakram 1) {dano_chakram1} + (Marca) {dano_marca1} + (Modificador) {modificador_atributo} = {dano_final_pa}')
        #print(f'Turno {i}: Ataque Secundário {rolagem_sa} com dano: (Chakram 2) {dano_chakram2} + (Marca) {dano_marca2} = {dano_final_sa} | Dano Total: {dano_final}')

        #Reset dos danos para não cumular
        dano_chakram1 = 0
        dano_chakram2 = 0
        dano_marca1 = 0
        dano_marca2 = 0
        dano_final_pa = 0
        dano_final_sa = 0
        dano_final = 0
        levou_dano = random.randint(1, 1000)

        if levou_dano <= 77:
          vantagem_grito = 0

    ################################################################################
    # Calcular a média dos resultados
    media_rolagens = sum(rolagens) / len(rolagens)
    media_dano = sum(ataques) / len(ataques)


    # Calculate and store the averages for this run
    if len(rolagens) > 0:
        media_rolagens = sum(rolagens) / len(rolagens)
        all_avg_rolagens.append(media_rolagens)

    if len(ataques) > 0:
        media_dano = sum(ataques) / len(ataques)
        all_avg_dano.append(media_dano)

    # Store the advantage status for this run
    all_advantage_attacks.extend(advantage_attacks_this_run)


# Calculate the overall average of the averages
overall_avg_rolagens = sum(all_avg_rolagens) / len(all_avg_rolagens) if len(all_avg_rolagens) > 0 else 0
overall_avg_dano = sum(all_avg_dano) / len(all_avg_dano) if len(all_avg_dano) > 0 else 0

# Calculate the percentage of attacks made with advantage
percentage_advantage = (sum(all_advantage_attacks) / len(all_advantage_attacks)) * 100 if len(all_advantage_attacks) > 0 else 0


print(f"Média geral das rolagens (ao longo de 100000 simulações): {overall_avg_rolagens}")
print(f"Média geral dos danos (ao longo de 100000 simulações): {overall_avg_dano}")
print(f"Porcentagem de ataques feitos com vantagem: {percentage_advantage:.2f}%")
```
